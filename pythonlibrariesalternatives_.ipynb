## Alternative Python Libraries for Common Data Tasks

Let's explore some unpopular but powerful Python libraries for data manipulation, visualization, and analysis. These libraries are excellent alternatives to the commonly used tools like pandas, matplotlib, and seaborn.

### Dataset Overview

We will begin by creating a small sample dataset with missing values and mixed data types, which we will use to show the functionality of these alternative libraries.

```python
# import libraries
import pandas as pd
import numpy as np
```

```python
# create sample data with missing values and mixed types
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah', 'Ian', 'Julia'],
    'Age': [25, np.nan, 35, 45, 29, 33, np.nan, 41, 38, 27],
    'Income': [50000, 60000, np.nan, 80000, 70000, 65000, np.nan, 72000, 69000, np.nan],
    'City': ['NY', 'LA', 'NY', 'Chicago', np.nan, 'Houston', 'Seattle', 'LA', np.nan, 'Chicago']
})
```

```python
# display the dataset
df
```

## 1. For data summary we will use skimpy instead of df.describe()

We often rely on df.describe(), which provides basic statistics, but doesn't show other information like missing values, most frequent values, etc. skimpy offers a more complete summary of your data with just one function call.

```python
# import skim from skimpy
import skimpy
from skimpy import skim

# get detailed data summary
skim(df)
```

### 2. Check missing data using missingno

It can be tedious to check missing data manually. A quick visual representation can be more insightful.


```python
# import missingno and matplotlib
import missingno as msno
import matplotlib.pyplot as plt

# visualize the missing data with a matrix plot
msno.matrix(df)
plt.show()
```

- Rows represent records indexed from 1 to 10.

- White bars indicate missing data.

- Black bars indicate non-missing data.

### 3. Create automated EDA report using sweetviz

Performing exploratory data analysis (EDA) manually can be time-consuming and error-prone however, sweetviz generates an interactive, detailed EDA report in a single command.

```python
# Import sweetviz
import sweetviz as sv

# Create the report
report = sv.analyze(df)

# Show the report as an HTML file
report.show_html('Sweetviz_Report.html')
```

### 4. Faster DataFrame operations using polars
For large datasets, pandas can become slow and inefficient. When working with big data, performance is crucial. A faster alernative is polars.

```python
# import polars
import polars as pl

# convert pandas dataFrame to polars dataFrame
df_polars = pl.from_pandas(df)

# perform aggregation: calculate the mean of Income and Age columns
result = df_polars.select([
    pl.col('Income').mean().alias('Average Income'),
    pl.col('Age').mean().alias('Average Age')
])

result
```

The above is a summary table showing the average income and age. f64 means the column datatypes are float 64.

### 5. Plotting using plotnine

Plotnine follows the grammar of graphics (like ggplot2 in R) and allows for concise, expressive code for generating plots.

```python
# import plotnine
from plotnine import ggplot, aes, geom_bar, theme_minimal

# create a bar plot showing the count of entries per city
ggplot(df, aes(x='City')) + \
    geom_bar(fill='skyblue') + \
    theme_minimal()
```

This plot is why I first learnt to code in R. The outcome is a stunning, visually appealing graphic, a nice break from the typical matplotlib defaults.

**Just a quick note:**

**The NaNs show up because we didn’t take care of the missing values in the 'City' column earlier. This is because that’s not what we’re focusing on right now. This post is all about demonstrating how different libraries in Python can create more polished visualizations right out of the box.**
